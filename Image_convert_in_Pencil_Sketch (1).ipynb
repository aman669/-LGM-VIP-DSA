#Import Libraries

#Read Image

```python
image = cv2.imread("horse-facts-name.jpg")
cv2_imshow(image)
```

#Gray Scale Image

```python
gray_image=cv2.imread("horse-facts-name.jpg",0)
cv2_imshow(gray_image)
```

#Invert the GrayScale Image

```python
inverte= 255-gray_image
cv2_imshow(inverte)
```

#Blur the Invert image by using Gaussian Function

```python
Blurr=cv2.GaussianBlur(inverte,(15,15),0)
cv2_imshow(Blurr)
```

#Pencil Sketch

```python
inverte_blurr= 255-Blurr
pencil_sketch = cv2.divide(gray_image,inverte_blurr,scale=256.0)
cv2_imshow(pencil_sketch)
```
